<h1>Вероятность. Событие. Частота событий.</h1>

<b>Теория вероятностей</b> - это математическая дисциплина, которая изучает случайные явления и события, их вероятности и свойства. Она предоставляет инструменты и методы для формального описания, моделирования и анализа случайных событий. Основные концепции теории вероятностей включают вероятностные пространства, случайные величины, распределения вероятностей, условные вероятности, независимость событий, а также законы, такие как закон больших чисел и центральная предельная теорема.<br>

<b>Вероятность</b> - числовая мера, числовая мера, отражающая уверенности в возможности наступления какого-либо события. Выражается в пределе [0;1] <br>


<b>Событие</b> - какой-либо исход, который может произойти в определенной ситуации. В теории вероятности описывается как набор исходов. <br>

<b>Частота событий</b> - это количество раз, которое некоторое событие происходит в определенной серии экспериментов или наблюдений. Например, если мы бросаем монету 100 раз и получаем 60 раз орел, то частота выпадения орла равна 60/100 или 0,6. <br>
<h2>Формула P=m\n. Противположные события.</h2>
События можно разделить на два вида:
<ol>
    <li>События, для которых можно точно определить вероятности их появления.</li>
    <li>События, для которых вероятности иъ появления удается вычислить лишь приблеженно.</li>
</ol>
Для первого вида можно использовать формулу:
$$P = \frac{m}{n}$$
Где:
<ul>
    <li>$n$ - размер полной группы несовместных событий.(кол-во событий)</li>
    <li>$m$ - число элементарных событий (из полной группы), благоприятных некорому исходу(для которого и вычисляется вероятность).</li>
</ul>
Основные характеристики:
<ul>
    <li>Образовывать полную группу событий, то есть кроме этих событий в этом эксперименте не могут появляться ни одно другое, они охватывают все возможные исходы.</li>
    <li>Каждое событие при проведении эксперимента возникает с равной вероятностью.(то есть равновероятны).</li>
    <li>Никаких два события из полной группы не могут произойти одновременно (другими словами, они дожны быть несовместными).</li>
</ul>
Противоположные события - события, которые охватывают все остальные исходы, которое не включает в себя событие $A$. Обозначается как $\overline{A}$
$$P(\overline{A})=1-P(A)$$
<i>Например:</i> <br>
Бросаются два игральных кубика и нужно нати вероятность того, что сумма очков не будет равна 11.
$\overline{A}$: сумма очков равна 11. Число благоприятных событий для этого исхода - 2: (5+6 ; 6+5)<br>
Тогда: $m=2$ ; $n=6 \cdot 6 = 36$ <br>
$P(\overline{A}) = \frac{m}{n}= \frac{2}{36} = \frac{1}{18}$<br>
$P(A) = 1 - P(\overline{A}) = 1 - \frac{1}{18} = \frac{17}{18}$

<h2>Зависимые/независимые события, условная вероятность, их произведение.</h2>
Два события называются независимыми, если появление одного из них не изменяет вероятности появления другого. В противном случае (появление одного события изменяет вероятность появления другого) события называются зависимыми.
Условная вероятность:<br>
Вероятность события $A$ при условии того, что событие $B$ произошло, называется условной верятностью и обозначается:
$$P(A|B)$
<h2>Произведение событий:</h2>
<h4>Произведение для независимых событий</h4>
Произведением $AB$ событий $A$ и $B$ называется событие, состоящее в том, что оба события произошли одновременно. Например: $A$ - идет дождь, $B$ - идет снег, то $AB$ - дождь со снегом.

Тогда: $$P(A \cdot B) = P(A) \cdot P(B)$$

<h4>Произведение для зависимых событий</h4>
Вероятность произведения двух зависимых событий равна произведению вероятности одного из них на условную вероятность другого. 

$$P(A \cdot B) = P(A) \cdot P(B|A)$$
Например:

![salam](https://cdn.discordapp.com/attachments/1204103200584761452/1227720155619131484/Screenshot_14.png?ex=66296e7f&is=6616f97f&hm=209de1fcd45e334490eed8b007bb51c837072801e468b92c5bb05af69a21c00c&)

$P(A) = \frac{3}{5}$ и $P(B) = \frac{2}{5}$ <br>

$P(A \cdot B) = ?$ - Сначало из урны взяли белый шар, а затем из оставшихся шаров вынули черный<br>
$P(A \cdot B) = P(A) \cdot P(B|A)$<br>
$P(A \cdot B) = \frac{3}{5} \cdot \frac{1}{2} = \frac{3}{10}$

<hr>

$P(B \cdot A) = ?$ - Сначало из урны взяли черный шар, а затем из оставшихся шаров вынули белый<br>
$P(B \cdot A) = P(B) \cdot P(A|B)$<br>
$P(A \cdot B) = \frac{2}{5} \cdot \frac{3}{4} = \frac{3}{10}$

Это означает что оба выражения равны и от перемены мест они равны: $P(AB)=P(BA)$

Исходя из данной задачи можно сделать вывод, что сумма для двух несовместных событий описывает формула:$$P(C)=P(A)+P(B)$$
или $$P(A+B)=P(A)+P(B)$$
Это означает в теории вероятности, такое событие $C$ есть тоже самое что и $C = A + B$, это значит, что данное событие $C$ интерприруется как появление события $A$ или события $B$. Или попадание дротиком или в мишень $A$ или в мишень $B$.
